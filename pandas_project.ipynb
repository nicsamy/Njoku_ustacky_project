## Using plot() function  to display the distribution of the products buy by female and men customers

```python
grouped = xyzdata.groupby(['Gender', 'Product line']).sum()['Quantity']
```

```python
grouped.unstack().plot(kind='bar', stacked=False)
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.title('Product Line and Quantity by Gender')
plt.show()
```

From the distribution above shows that Female purchase more fashion accessories than men and the men purchase more health and beauty products than female. From thebar chart above the female also, purchase more food and beverages than the men.

```python
# Plot the data as a bar plot
plt.figure(figsize=(11, 6))
sns.barplot(x=grouped.index.get_level_values(1), y=grouped.values, hue=grouped.index.get_level_values(0))
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.title('Product Line and Quantity by Gender')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
product_lines = ['Electronic accessories', 'Fashion accessories', 'Food and beverages','Health and beauty','Home and lifestyle','Sport and travel']
genders = ['Male','Female']
df_filtered =xyzdata[xyzdata['Product line'].isin(product_lines) & xyzdata['Gender'].isin(genders)]

# Plot the distribution of the quantity column for each product line and gender
for product_line in product_lines:
    for gender in genders:
        data = df_filtered[(df_filtered['Product line'] == product_line) & (df_filtered['Gender'] == gender)]['Quantity']
        sns.kdeplot(data, label='{} - {}'.format(product_line, gender))

plt.xlabel('Quantity')
plt.title('Distribution of Quantity by Product Line and Gender')
plt.legend()
plt.show()

```
